#Aggregate statisctis(Group by)

```python
# #refresh to the modified dataframe
# df = pd.read_csv('modified.csv')

# #average according to the group, in this case it is type 1
# df.groupby(['Type 1'])
# df.head()
```

```python
#refresh to the modified dataframe
df = pd.read_csv('modified.csv')

#average according to the group, in this case it is type 1
df.groupby(['Type 1']).mean()
```

```python
#sorting according to the mean on the column/feature "Attack"
df.groupby(['Type 1']).mean().sort_values("Attack", ascending = False)
```

```python
df.groupby(['Type 1']).count()
```

```python
#refresh to the modified dataframe
df = pd.read_csv('modified.csv')

df['count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```

```python

df.groupby(['Type 1']).count()['count']
```

#working with large amounts of data

```python
#Creating a new dataframe with the same columns
new_df = pd.DataFrame(columns=df.columns)

#to finalise an output dataframe that is a result of all the above operation
for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
```
